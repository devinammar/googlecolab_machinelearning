```python
# Load Dataset

# Load dataset WDBC
# Kita kasih nama kolom karena file .data aslinya gapunya header
cols = ['ID', 'Diagnosis'] + [f'Feature_{i}' for i in range(1, 31)]
df = pd.read_csv('wdbc.data', header=None, names=cols)

# Tampilkan 5 data teratas buat mastiin udah masuk
df.head()
```

```python
# --- B. Analisis Missing Value ---
print(f"Total Missing Value: {df.isnull().sum().sum()}")

# --- C. Categorical Encoding ---
# Ubah M (Malignant) jadi 1 dan B (Benign) jadi 0
df['Diagnosis'] = df['Diagnosis'].map({'M': 1, 'B': 0})
print("Encoding: Diagnosis berhasil diubah ke angka (M=1, B=0)")

# --- D. Feature Scaling ---
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Ambil fitur aja (ID dan Diagnosis jangan di-scale)
fitur_cols = df.columns[2:]
df[fitur_cols] = scaler.fit_transform(df[fitur_cols])

print("Scaling: Semua fitur sudah di-standarisasi dengan StandardScaler")
```
